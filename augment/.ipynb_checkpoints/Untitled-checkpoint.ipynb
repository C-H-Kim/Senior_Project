```python
import os
from glob import glob
from PIL import Image
```

```python
basic_path = '/home/chkim/SeniorProject/augment/merge_data/'
dir_list = os.listdir(basic_path)
len(dir_list)
```

```python
for i in range(len(dir_list)):
    print(f"[INFO] {dir_list[i]} started")
    images = glob(basic_path + dir_list[i] + '/*.jpg')
    for im in images:
        img = Image.open(im)
        w, h = img.size
        if (w != 224) or (h != 224):
            print(im)
    print(f"[INFO] {dir_list[i]} done")
```
