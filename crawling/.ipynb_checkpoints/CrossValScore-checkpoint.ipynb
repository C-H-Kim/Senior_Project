```python
import time
from PIL import Image

features = []
labels = []

start = time.time()

for im in images:
    labels.append(im[len("./crawl_image_duplicate/"):-len('_0000.jpg')])
    im = mh.imread(im)
    im = mh.colors.rgb2grey(im, dtype=np.uint8)
    features.append(mh.features.haralick(im).ravel())

print(f'fit time : {time.time() - start}')
```

```python
features = np.array(features)
features
```

```python
def selectimage(n, m, dists, images):
    image_position = dists[n].argsort()[m]
    image = mh.imread(images[image_position])
    return image
```

```python
def plotImages(n):
    fig, ax = plt.subplots(5, 5, figsize = (15,5))
    
    for i in range(25):
        ax[i].imshow(selectimage(n,i, dists, images))
        ax[i].set_xticks([])
        ax[i].set_yticks([])
        
    plt.show()
```
